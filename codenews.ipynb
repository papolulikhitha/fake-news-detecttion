```python
import pandas as pd

fake = pd.read_csv(r"C:\Users\user\Desktop\datasets\News _dataset\Fake.csv")
true = pd.read_csv(r"C:\Users\user\Desktop\datasets\News _dataset\True.csv")

news = input("Enter news: ").lower()

fake_count = (
    fake["title"].str.lower().str.contains(news, na=False).sum()
    + fake["text"].str.lower().str.contains(news, na=False).sum()
)

true_count = (
    true["title"].str.lower().str.contains(news, na=False).sum()
    + true["text"].str.lower().str.contains(news, na=False).sum()
)

if fake_count > true_count:
    print("Prediction: FAKE")
elif true_count > fake_count:
    print("Prediction: TRUE")
elif fake_count == 0 and true_count == 0:
    print("News not found")
else:
    print("Cannot determine")
```
